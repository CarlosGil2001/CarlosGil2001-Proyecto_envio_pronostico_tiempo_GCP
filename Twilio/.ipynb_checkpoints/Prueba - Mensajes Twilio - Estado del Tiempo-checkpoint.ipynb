## Armando la URL

```python
# Definir la ciudad al que vamos a obtener el estado del tiempo
query = 'Lima'
# API que nos proporciona WEATHER API para obtener el estado del tiempo
api_key = API_KEY_WAPI

# Obtenemos de 2 días
url_estado_clima ='http://api.weatherapi.com/v1/forecast.json?key='+api_key+'&q='+query+'&days=2&aqi=no&alerts=no'
url_estado_clima
```

## Crear Dataframe

```python
# Función que recorre y extrae los valores de interés del JSON - API
def get_forecast(response,i):
    fecha = response_clima['forecast']['forecastday'][0]['hour'][i]['time'].split()[0]
    hora = int(response_clima['forecast']['forecastday'][0]['hour'][i]['time'].split()[1].split(':')[0])
    condicion = response_clima['forecast']['forecastday'][0]['hour'][i]['condition']['text']
    tempe_cel = float(response_clima['forecast']['forecastday'][0]['hour'][i]['temp_c'])
    tempe_fah=float(response_clima['forecast']['forecastday'][0]['hour'][i]['temp_f'])
    por_humedad = float(response_clima['forecast']['forecastday'][0]['hour'][i]['humidity']/100)
    por_nubosidad=float(response_clima['forecast']['forecastday'][0]['hour'][i]['cloud']/100)
    rain = response_clima['forecast']['forecastday'][0]['hour'][i]['will_it_rain']
    prob_rain = response_clima['forecast']['forecastday'][0]['hour'][i]['chance_of_rain']
    
    return fecha,hora,condicion,tempe_cel,tempe_fah,por_humedad,por_nubosidad,rain,prob_rain
```

```python
# Recorrer los valores del estado del tiempo
datos = []

for i in tqdm(range(len(response_clima['forecast']['forecastday'][0]['hour'])),colour = 'blue'):
    # agregamos valores al array
    datos.append(get_forecast(response_clima,i))
```

```python
# Consultamos el array con los datos del estado del tiempo
datos
```

```python
# Definimos los nombre de las columnas de DF
columns = ['Fecha','Hora','Condicion','Temperatura_Cel','Temperatura_Fah','%_Humedad','%_Nubosidad','Lluvia','Prob_Lluvia']

# Crear el DF
df_clima = pd.DataFrame(datos,columns=columns)
df_clima = df_clima.sort_values(by = 'Hora',ascending = True)
df_clima
```

```python
# Información adicional necesaria
df_info_ad = pd.DataFrame({
    '%Nubosidad': [round(df_clima['%_Nubosidad'].mean()*100, 0)],
    '%Humedad': [round(df_clima['%_Humedad'].mean()*100, 0)],
    'Temperatura': [round(df_clima['Temperatura_Cel'].mean(), 1)],
    'Condicion': [df_clima['Condicion'].mode()[0]]
})
df_info_ad = df_info_ad.reset_index(drop=True)
df_info_ad
print(df_info_ad.to_string(index=False))
```

```python
# Filtramos los datos para cuando la columna LLUVIA sea 1 y la hora esté entre las 6 a.m y 23 p.m
df_clean =  df_clima[(df_clima['Lluvia']==1) & (df_clima['Hora']>6) & (df_clima['Hora']<= 23)]
df_rain = df_clean[['Hora','Condicion','Temperatura_Cel']]
df_rain.set_index('Hora', inplace = True)

# Mostra el DF limpio
df_rain
```
